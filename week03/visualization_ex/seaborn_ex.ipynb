```python
import seaborn as sns

sns.lineplot(x = [1, 3, 2, 4], y = [0.7, 0.2, 0.1, 0.05])
```

```python
sns.barplot(x = [1, 3, 2, 4], y = [0.7, 0.2, 0.1, 0.05])
```

```python
import matplotlib.pyplot as plt

sns.barplot(x = [1, 3, 2, 4], y = [0.7, 0.2, 0.1, 0.05])
plt.title("Bar Plot")
plt.show()
```

```python
sns.barplot(x = [1, 3, 2, 4], y = [0.7, 0.2, 0.1, 0.05])
plt.xlabel("X Label")
plt.ylabel("Y Label")

plt.show()
```

```python
sns.lineplot(x = [1, 3, 2, 4], y = [4, 3, 2, 1])
plt.ylim(2, 3)

plt.show()
```

```python
plt.figure(figsize = (20, 10))
sns.lineplot(x = [1, 3, 2, 4], y = [4, 3, 2, 1])

plt.show()
```
